```python
x = [1,2,3,4]
y = [11,22,33,44]
plt.ylabel('Quantity of Milk used in Litres')
plt.xlabel('No. of weeks')

plt.plot(x,y);
```

```python
# 1st method
fig = plt.figure()  # create a figure
ax = fig.add_subplot() # adds some axes
plt.show()
```

```python
## 2nd method
fig = plt.figure() ## create a figure
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# 3rd method (recommended)
fig, ax = plt.subplots()
ax.plot(x,y);  # add some data
```
